```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_excel("data.xlsx", index_col='Serial_no')
df.head(3)
```

```python
df = df.dropna(axis=1)
df = df.loc[:, df.columns != 'Q_1']
df.head(3)
```

```python
fig = plt.figure(figsize=(22, 40))
k = 1
for i in df.drop(['Q_3', 'Q_4', 'Q_5', 'Q_7', 'Q_18', 'Q_19'], axis=1).columns:
    plt.subplot(13, 4, k)
    df[i].value_counts().plot(kind='pie', xlabel=i)
    k += 1
plt.tight_layout()
plt.show() 
```
